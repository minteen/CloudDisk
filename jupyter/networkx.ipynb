## 基础操作

在networkx中，一个节点可以是任意hash对象（除None对象外），一条边也可以关联任意的对象，包括但不限于文本字符串、图像或另一个图。

### 创建图

- 四种内置图类：
	- nx.Graph() 无向图
	- nx.DiGraph() 有向图
	- nx.MultiGraph() 多重无向图
	- nx.MultiDigraph() 多重有向图

```python
import networkx as nx

# 创建图
G = nx.DiGraph(info="Learning Networkx")
```

### 添加/删除/查找 节点/边

如果添加的节点或边已经存在，不会报错；
如果添加的边中包含未创建的节点，会自动创建

- `G.add_node('a')`
- `G.add_nodes_from([1, 'a', ])`
- `G.add_node(H)` 将子图H加入G中
---
- `G.remove_node(1)`
- `G.remove_nodes_form([1, 2])`
- `G.remove_edge(1, 2)`
- `G.remove_edges_form([(1, 2), (3, 4)])`
---
- `G.add_edge(1, 2)`
- `G.add_edges_from([(1, 2),(3, 5)])`
- `G.add_edges_from(H.edges())`
- `G.add_weighted_edges_from([(1,2,0.125), (1,3,0.75)])`
---
- `G.has_node(1)`
- `G.has_edge(1, 2)`

```python


# 添加节点和边
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
G.add_edges_from([("A", "B"), ("A", "C"), ("E", "A"),("B", "D"), ("D", "F"), ("E", "G")])
print(G.nodes())
print(G.edges())
nx.draw(G, with_labels=True)

# h is a subGraph of g
# h = nx.random_graphs.erdos_renyi_graph(10, 0.3, directed=True)
# g.add_node(h)
```
